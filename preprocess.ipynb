```python
#preprocess step

import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
from pandas.api.types import CategoricalDtype


# FUNCTION TO HANDLE OUTLIERS
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df


# MOVIES DATASET
movies_df = pd.read_csv(r"C:\Users\nh013\Desktop\movie recommendation system dataset\movies.csv")


#RATINGS DATASET
ratings_df = pd.read_csv(r"C:\Users\nh013\Desktop\movie recommendation system dataset\ratings.csv")


# MOVIES DATAFRAME FEATURE EXTRACT
movies_df = movies_df[['movieId', 'title', 'genres']]



# FEATURE EXTRACT RATING DATAFRAME 
ratings_df = ratings_df[['userId', 'movieId', 'rating', 'timestamp']]



# CHECK MISSING VALUES
movies_df.dropna(inplace=True)
ratings_df.dropna(inplace=True)



# HANDLE OUTLIERS
movies_df = handle_outliers(movies_df, 'movieId')
ratings_df = handle_outliers(ratings_df, 'rating')

#NORMALIZE AND SCLAE NUMERIC COLUMN TO MOVIES DATAFRAME 
numeric_cols_movies = ['movieId']
scaler = MinMaxScaler()
movies_df[numeric_cols_movies] = scaler.fit_transform(movies_df[numeric_cols_movies])


# NORMALIZE AND SCLAE NUMERIC COLUMN TO RATING DATAFRAME
numeric_cols_ratings = ['rating']
scaler = StandardScaler()
ratings_df[numeric_cols_ratings] = scaler.fit_transform(ratings_df[numeric_cols_ratings])



# CONVERT CATEGORICAL VALUES TO NUMERICAL TO MOVIES DTAFRAME 
categorical_cols_movies = ['title', 'genres']
for col in categorical_cols_movies:
    cat_type = CategoricalDtype(categories=movies_df[col].unique(), ordered=True)
    movies_df[col] = movies_df[col].astype(cat_type).cat.codes

    
    
# CONVERT CATEGORICAL VALUES TO NUMERICAL TO RATING DATAFRAME
categorical_cols_ratings = ['userId']
for col in categorical_cols_ratings:
    cat_type = CategoricalDtype(categories=ratings_df[col].unique(), ordered=True)
    ratings_df[col] = ratings_df[col].astype(cat_type).cat.codes

    
    
#ONE HOT ENCODING TO MOVIES DATAFRAME
movies_df_encoded = pd.get_dummies(movies_df, columns=categorical_cols_movies, sparse=True)

#ONE HOT ENCODING TO RATING DATAFRAME
ratings_df_encoded = pd.get_dummies(ratings_df, columns=categorical_cols_ratings, sparse=True)



print("Movies DataFrame:")
print(movies_df_encoded.head())

print("Ratings DataFrame:")
print(ratings_df_encoded.head())
```
